### **Location Normalization Ideas**

* Through manual inspection, the majority of the location is on the city
level. However, some of the locations are on the country level. 


*   We can take the lowest level location as our final location.
(e.g. Arlington,VA,USA ----> Arlington) # split into city, state/province, country, in separate columns, with country highest priority

*   Do a split by "," and take the first element from the list

*   Also noticed some inconsistency for some foreign city name
(e.g. Xi'an and Xi an) # use string with normalization and whitespace stripped to match all variants, and then map all these to a standardized form

*   Create a dictionary with the city name with all lower letter and symbol,
space(s) removed as key. The standardize city name as value.

*   Literate through the location list, delete all symbols and spaces, and 
change the location to all lower letters. Matched the changed location to the
dictionary. If the key exists, assign value to that location. If the key
does not exists, create a new dictionary entry with the standardize value.

## Python Functions Used For the Report


### Get the count of distinct values and graph a histogram


```
def countDistinct(colname):
  ### if the columns have all distinct value return True
  if len(conferences) == len(set(conferences[colname])):
    return "Data in this column have all distinct value"
  sql_string_build = "select "+ colname +", count(" + colname + ") as num_ids from gcp_cset_mag.ConferenceInstances group by " + colname +" order by num_ids desc"
  temp_df = pandas.read_gbq(sql_string_build, dialect="standard", project_id="gcp-cset-projects")
  #### change the temp df colname column to string (for graphing the barchart)
  temp_str = []
  for i in range(len(temp_df)):
    temp_str.append(str(temp_df[colname][i]))
  temp_df[colname] = temp_str
  #### Close to have all distinct value return True
  if temp_df['num_ids'][0] <= 5:
    return "Data in this column have almost all distinct value"
  ### else return the dataframe
 # print(temp_df.dtypes)
 # barchart = px.bar(
     # data_frame = temp_df,
      #x = colname,
      #y = "num_ids"
 # )
 ### Those column has null value as their most frequent element, I decided to remove it for a better barchart 
  if colname == 'OfficialUrl' or colname == 'StartDate' or colname == 'EndDate'or colname == 'AbstractRegistrationDate'or colname == 'SubmissionDeadlineDate'or \
  colname == 'NotificationDueDate'or colname == 'FinalVersionDueDate' or colname == 'PaperCount' or colname == 'PaperFamilyCount' or colname == 'CitationCount' \
  or colname == 'Latitude' or colname == 'Longitude':
    temp_df = temp_df.drop(0)

  if colname == 'ConferenceSeriesId' or colname == 'StartDate' or colname == 'EndDate'or colname == 'AbstractRegistrationDate'or colname == 'SubmissionDeadlineDate' \
  or colname == 'NotificationDueDate' or colname == 'FinalVersionDueDate'or colname == 'Latitude' or colname == 'Longitude' or colname == 'import_time' or colname == 'CitationCount':
    fig = go.Figure([go.Bar(x = [i for i in range(len(temp_df))], y = temp_df['num_ids'],  marker_color =  '#123fff', hovertext = temp_df[colname])])
  else:
    fig = go.Figure([go.Bar(x = temp_df[colname], y = temp_df['num_ids'], marker_color =  '#123fff', hovertext = temp_df[colname])])
  pio.show(fig)
  #pio.show(barchart)
  return temp_df
```

### Check if the column contains na, n/a, none, TBD and other that is encoded as null


```
def checknull(colname):
  str_to_return = ''
  if "na" in set(conferences[colname]):
    str_to_return += "have na "
  if "n/a" in set(conferences[colname]):
    str_to_return += "have n/a "
  if "NA" in set(conferences[colname]):
    str_to_return += "have NA "
  if "N/A" in set(conferences[colname]):
    str_to_return += "have N/A "
  if "None" in set(conferences[colname]):
    str_to_return += "have None "
  if "none" in set(conferences[colname]):
    str_to_return += "have none "
  if "Null" in set(conferences[colname]):
    str_to_return += "have Null "
  if "null" in set(conferences[colname]):
    str_to_return += "have null "
  if None in set(conferences[colname]):
    str_to_return += "have NoneType "
  if "TBD" in set(conferences[colname]):
    str_to_return += "have TBD "
  sql_string_build = "select count(" + colname + ") as num_ids from gcp_cset_mag.ConferenceInstances where regexp_contains( CAST(" + colname+ " AS STRING) , r\"^\s*$\")"
  temp_df = pandas.read_gbq(sql_string_build, dialect="standard", project_id="gcp-cset-projects")
  if temp_df['num_ids'][0] > 0:
    str_to_return += "have empty space(s)"


  if str_to_return:
    return str_to_return
  return "There is no null value detected by the function, but need a manual check"
```

### Bar Chart for numerical Data
